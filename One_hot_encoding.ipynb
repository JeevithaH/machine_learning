```python
import pandas as pd
```

```python
url= 'https://github.com/JeevithaH/machine_learning/raw/main/price.csv'
```

```python
df=pd.read_csv(url,encoding='unicode_escape')
df
```

```python
dummies=pd.get_dummies(df.town)
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
#drop orignal town column
#drop west windsor for removing dummy varible trap(remove multicoleanearity)
final=merged.drop(['town','west windsor'],axis ='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#as we are detecting the price of the area this price will be dependent variable ie.,Y
#rest will be x independent variable
x=final.drop('price',axis="columns")
```

```python
x
```

```python
y=final.price
```

```python
y
```

```python
#for training
model.fit(x,y)
```

```python
#price of 2800 sq feet area in robinsville
model.predict([[2800,0,1]])
```

```python
model.predict([[5600,1,0]])
```

```python
# for price in west windsor make both 2 towns as 0
model.predict([[4000,0,0]])
```

```python
#accuracy of model
model.score(x,y)
#this model iis 34% accurate
```
